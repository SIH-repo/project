```python
import pandas as pd
file_path = '/content/drive/My Drive/cve.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python

df.dropna(subset=['summary', 'cvss'], inplace=True)
X = df['summary']
y = df['cvss']
vectorizer = TfidfVectorizer(max_features=1000, stop_words='english')
X_vectorized = vectorizer.fit_transform(X).toarray()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_vectorized, y, test_size=0.2, random_state=42)
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
df['predicted_cvss'] = model.predict(vectorizer.transform(df['summary']).toarray())
```

```python
def format_for_email(row):
    return {
        "CVE ID": row['Unnamed: 0'],
        "CWE Code": f"{row['cwe_code']} - {row['cwe_name']}",
        "Summary": row['summary'],
        "Predicted CVSS Score": row['predicted_cvss'] if pd.isna(row['cvss']) else row['cvss'],
        "Impact": {
            "Confidentiality": row['impact_confidentiality'],
            "Integrity": row['impact_integrity'],
            "Availability": row['impact_availability']
        }
    }
categorized_data = df.apply(format_for_email, axis=1).tolist()
formatted_df = pd.DataFrame(categorized_data)
formatted_df.to_csv('/content/drive/My Drive/formatted_cve_data.csv', index=False)
formatted_df.head()
```

```python
def format_for_email(row):
    return {
        "CVE ID": row['Unnamed: 0'],
        "CWE Code": f"{row['cwe_code']} - {row['cwe_name']}",
        "Summary": row['summary'],
        "Predicted CVSS Score": row['predicted_cvss'] if pd.isna(row['cvss']) else row['cvss'],
        "Impact": {
            "Confidentiality": row['impact_confidentiality'],
            "Integrity": row['impact_integrity'],
            "Availability": row['impact_availability']
        }
    }
categorized_data = df.apply(format_for_email, axis=1).tolist()
formatted_df = pd.DataFrame(categorized_data)
formatted_df.to_csv('/content/drive/My Drive/formatted_cve_data.csv', index=False)
formatted_df.head()
```

```python
def determine_severity(cvss):
    if cvss >= 9.0:
        return "Critical"
    elif cvss >= 7.0:
        return "High"
    elif cvss >= 4.0:
        return "Medium"
    else:
        return "Low"
def determine_mitigation(severity):
    if severity == "Critical":
        return "Immediate patch is required."
    elif severity == "High":
        return "Apply security updates as soon as possible."
    elif severity == "Medium":
        return "Schedule a security patch during maintenance."
    else:
        return "Monitor the system and apply a patch if necessary."
def format_for_email(row):
    cvss_score = row['predicted_cvss'] if pd.isna(row['cvss']) else row['cvss']
    severity = determine_severity(cvss_score)
    mitigation_strategy = determine_mitigation(severity)

    return {
        "CVE ID": row['Unnamed: 0'],
        "CWE Code": f"{row['cwe_code']} - {row['cwe_name']}",
        "Summary": row['summary'],
        "Predicted CVSS Score": cvss_score,
        "Severity Level": severity,
        "Impact": {
            "Confidentiality": row['impact_confidentiality'],
            "Integrity": row['impact_integrity'],
            "Availability": row['impact_availability']
        },
        "Mitigation Strategy": mitigation_strategy
    }
categorized_data = df.apply(format_for_email, axis=1).tolist()
formatted_df = pd.DataFrame(categorized_data)
formatted_df.to_csv('/content/drive/My Drive/formatted_cve_data.csv', index=False)
formatted_df.head()
```
